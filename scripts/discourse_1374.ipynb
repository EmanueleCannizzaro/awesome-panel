```python
import os
import geopandas as gpd
import pandas as pd
import panel as pn
import param
pn.extension()


class Streets(param.Parameterized):
    mun = {'MUN1':'Mun1', 'MUN2':'Mun2'} # selecting from the selector municipality to get the path (select MUN1 to get path G:\\Test\\Mun1)
    
    select_mun = param.Selector(default=list(mun.keys())[0], objects=list(mun.keys())) # selector for municipality
    files = param.MultiFileSelector() # to select final folder in the path
    

    view = param.Parameter()

    def __init__(self, **params):
        super().__init__(**params)
        self.folder = 'G:\\Test'                   #change to existing starting folder
        self.view = self._create_view()
        
    @param.depends('select_mun', watch=True)
    def _update_file(self):
               
        self.mun = mun[self.select_mun]
        self.munfolder = os.path.join(self.folder, self.mun)
        self.outputfilename = self.mun + "_control.xlsx"
        self.outputpat = os.path.join(self.munfolder, self.outputfilename)
        
        self.param['files'].path = self.munfolder   # file selector is not updated. Showing default path on computer. It is not updated to the folder G:\\Test\\Mun1
        
        print('files', self.files)
        
    
    def _create_view(self):
        
        panel_left = pn.Param(
            self, 
            parameters=["select_mun", "files"],
            widgets={
                'select_mun': {'widget_type': pn.widgets.Select}, 
                'files': {'widget_type': pn.widgets.FileSelector}}
            )
        
        print('select_mun = ',self.select_mun)
        print('files = ',self.files)
        
        return pn.Column(panel_left)
    
Streets().view.servable()
```
