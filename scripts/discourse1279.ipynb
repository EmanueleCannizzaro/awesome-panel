```python
import holoviews as hv
import hvplot.pandas
import pandas as pd
import panel as pn
import numpy as np

panel = pn.Column()
```

```python
df = pd.DataFrame({
    "x": [1, 1, 2, 2, 3, 3],
    "g": ["One", "Two", "Three"] * 2,
    "a": np.random.rand(6),
    "b": np.random.rand(6),
})
plt = df.hvplot.bar(
    y=["a", "b"], stacked=True, groupby="g",
    legend=False,
)
#plt = pn.Pane(plt, linked_axes=False)
panel.append(plt)
```

```python
df = pd.DataFrame({
    "xx": [1, 1, 2, 2, 3, 3],
    "gg": ["One1", "Two2", "Three3"] * 2,
    "cc": np.random.rand(6),
    "dd": np.random.rand(6),
})
plt = df.hvplot.bar(
    y=["cc", "dd"], stacked=True, groupby="gg",
    legend=False,
)
#plt = pn.Pane(plt, linked_axes=False)
panel.append(plt)
```
